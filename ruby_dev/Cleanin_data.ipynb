```python
covid_death_gender['Age Ranges'] = covid_death_gender.loc(['Age Ranges'('0-17 years')])
```

```python
def ages(row):
    if row['Age Group'] == '0-17 years':
        return 'Child'
    elif row['Age Group'] == '1-4 years':
        return 'Child'
    
    elif row['Age Group'] == '5-14 years':
        return 'Child'
    
    elif row['Age Group'] == '18-29 years':
        return 'Adults'
    
    elif row['Age Group'] == '30-39 years':
        return 'Adults'
    
    elif row['Age Group'] == '40-49 years':
        return 'Adults'
    
    elif row['Age Group'] == '50-64 years':
        return 'Senior'
    
    elif row['Age Group'] == '65-74 years':
        return 'Senior'
    
    elif row['Age Group'] == '75-84 years':
        return 'Senior' 
    
```

```python
covid_death_gender['Ages'] = covid_death_gender.apply (lambda row: ages(row), axis=1)
covid_death_gender
```
