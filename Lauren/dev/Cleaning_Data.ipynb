```python
#drop unneeded comlumns
deaths_df = deaths_df.drop(["Data As Of", "Start Date", "End Date", "Year", "Month", "MMWR Week", "Total Deaths", "Footnote", "Age Group"], axis=1)
```

```python
#drop unneeded rows
deaths_df = deaths_df.loc[(deaths_df["HHS Region"] != "United States") & (deaths_df["Group"] == "By Week") & (deaths_df ["Race and Hispanic Origin Group"] != "Unknown")]
```

```python
#drop group column
deaths_df = deaths_df.drop(["Group"], axis=1)

#rename columns
deaths_df = deaths_df.rename(columns={"Week-Ending Date": "Week", "HHS Region": "HHS_Region", "Race and Hispanic Origin Group": "Race", "COVID-19 Deaths": "COVID_Deaths"})
```

```python
#convert week to datetime
deaths_df["Week"] = pd.to_datetime(deaths_df["Week"])
```

```python
#group data
#get total deaths, by race, in HHS region 1 by week
deaths_df_1 = deaths_df[deaths_df["HHS_Region"] == "01"]
deaths_df_1 = deaths_df_1.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 2 by week
deaths_df_2 = deaths_df[deaths_df["HHS_Region"] == "02"]
deaths_df_2 = deaths_df_2.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 3 by week
deaths_df_3 = deaths_df[deaths_df["HHS_Region"] == "03"]
deaths_df_3 = deaths_df_3.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 4 by week
deaths_df_4 = deaths_df[deaths_df["HHS_Region"] == "04"]
deaths_df_4 = deaths_df_4.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 5 by week
deaths_df_5 = deaths_df[deaths_df["HHS_Region"] == "05"]
deaths_df_5 = deaths_df_5.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 6 by week
deaths_df_6 = deaths_df[deaths_df["HHS_Region"] == "06"]
deaths_df_6 = deaths_df_6.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 7 by week
deaths_df_7 = deaths_df[deaths_df["HHS_Region"] == "07"]
deaths_df_7 = deaths_df_7.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 8 by week
deaths_df_8 = deaths_df[deaths_df["HHS_Region"] == "08"]
deaths_df_8 = deaths_df_8.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 9 by week
deaths_df_9 = deaths_df[deaths_df["HHS_Region"] == "09"]
deaths_df_9 = deaths_df_9.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)

#get total deaths, by race, in HHS region 10 by week
deaths_df_10 = deaths_df[deaths_df["HHS_Region"] == "10"]
deaths_df_10 = deaths_df_10.groupby(["Week", "Race"]).sum()["COVID_Deaths"].astype(int)
```

```python
# create HHS region data
groups = {
    "01": ["Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island", "Vermont"],
    "02": ["New Jersey", "New York", "Puerto Rico", "Virgin Islands"],
    "03": ["Delaware", "District of Columbia", "Maryland", "Pennsylvania", "Virginia", "West Virginia"],
    "04": ["Alabama", "Florida", "Georgia", "Kentucky", "Mississippi", "North Carolina", "South Carolina", "Tennessee"],
    "05": ["Illinois", "Indiana", "Michigan", "Minnesota", "Ohio", "Wisconsin"],
    "06": ["Arkansas", "Louisiana", "New Mexico", "Oklahoma", "Texas"],
    "07": ["Iowa", "Kansas", "Missouri", "Nebraska"],
    "08": ["Colorado", "Montana", "North Dakota", "South Dakota", "Utah", "Wyoming"],
    "09": ["Arizona", "California", "Hawaii", "Nevada"],
    "10": ["Alaska", "Idaho", "Oregon", "Washington"]
}

##edit week data to reflect when the vaccination would go into effect - aka 2 weeks after distro
weeks = {
    "1/2/21": "12/14/2020",
    "1/9/21" :'12/21/2020',
    "1/16/21" : '12/28/2020',
    "1/23/21" :'12/28/2020',
    "1/30/21" : '01/11/2021',
    "2/6/21" : '01/18/2021',
    "2/13/21" : '01/25/2021',
    "2/20/21" : '02/01/2021',
    "2/27/21" :'02/08/2021',
    "3/6/21" : '02/15/2021',
    "3/13/21" : '02/22/2021',
    "3/20/21" : '03/01/2021',
    "3/27/21" : '03/08/2021',
    "4/3/21" : '03/15/2021',
    "4/10/21" : '03/22/2021',
    "4/17/21" :'03/29/2021',
    "N/A" : '04/05/2021',
    "N/A" : '04/12/2021',
    "N/A" : '04/19/2021',
    "N/A" : '04/26/2021'
}
```

```python
#clean JJ data
#add HHS region data to JJ DF
JJ_df["HHS_Region"] = (
    JJ_df["Jurisdiction"]
    .apply(lambda x: [k for k in groups.keys() if x in groups[k]])
    .str[0]
    .fillna("Other")
)

#edit weeks
JJ_df["Week"] = (
    JJ_df["Week of Allocations"]
    .apply(lambda x: [k for k in weeks.keys() if x in weeks[k]])
    .str[0]
    .fillna("Other")
)

#rename columns
JJ_df = JJ_df.rename(columns={"1st Dose Allocations": "Vaccinations"})

#drop n/a weeks and columns
JJ_df = JJ_df.drop(["Jurisdiction", "Week of Allocations"], axis = 1)
JJ_df = JJ_df.loc[(JJ_df["Week"] != "N/A") & (JJ_df["Week"] != "Other")]
```

```python
#convert week to datetime
JJ_df["Week"] = pd.to_datetime(JJ_df["Week"])
```

```python
#get total JnJ vaccines distributed
Total_JJ = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 1 by week
JJ_df_1 = JJ_df[JJ_df["HHS_Region"] == "01"]
JJ_df_1 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 2 by week
JJ_df_2 = JJ_df[JJ_df["HHS_Region"] == "02"]
JJ_df_2 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 3 by week
JJ_df_3 = JJ_df[JJ_df["HHS_Region"] == "03"]
JJ_df_3 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 4 by week
JJ_df_4 = JJ_df[JJ_df["HHS_Region"] == "04"]
JJ_df_4 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 5 by week
JJ_df_5 = JJ_df[JJ_df["HHS_Region"] == "05"]
JJ_df_5 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 6 by week
JJ_df_6 = JJ_df[JJ_df["HHS_Region"] == "06"]
JJ_df_6 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 7 by week
JJ_df_7 = JJ_df[JJ_df["HHS_Region"] == "07"]
JJ_df_7 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 8 by week
JJ_df_8 = JJ_df[JJ_df["HHS_Region"] == "08"]
JJ_df_8 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 9 by week
JJ_df_9 = JJ_df[JJ_df["HHS_Region"] == "09"]
JJ_df_9 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)

#get total JnJ vaccines distributed in HHS region 10 by week
JJ_df_10 = JJ_df[JJ_df["HHS_Region"] == "10"]
JJ_df_10 = JJ_df.groupby(["Week"]).sum()["Vaccinations"].astype(int)
```
