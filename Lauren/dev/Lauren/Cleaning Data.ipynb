CLEANING DEATHS DATA

```python
#read in deaths data
deaths_df = pd.read_csv("Resources/Weekly_deaths.csv")
```

```python
deaths_df
```

```python
deaths_df.dtypes
```

```python
#drop unneeded comlumns
deaths_df = deaths_df.drop(["Data As Of", "Start Date", "End Date", "Year", "Month", "MMWR Week", "Total Deaths", "Footnote", "Age Group"], axis=1)
```

```python
#drop unneeded rows
deaths_df = deaths_df.loc[(deaths_df["HHS Region"] != "United States") & (deaths_df["Group"] == "By Week") & (deaths_df ["Race and Hispanic Origin Group"] != "Unknown")]
```

```python
deaths_df
```

```python
deaths_df.dtypes
```

```python
#convert week to datetime
deaths_df["Week-Ending Date"] = pd.to_datetime(deaths_df["Week-Ending Date"])
```

```python
deaths_df
```

```python
#use get dummies to create variables for race
deaths_df = pd.get_dummies(deaths_df, columns=["Race and Hispanic Origin Group"])
deaths_df
```
