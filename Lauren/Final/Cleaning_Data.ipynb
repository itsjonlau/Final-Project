```python
#import dependencies
import pandas as pd
```

```python
#read in deaths data
deaths_df = pd.read_csv("Resources/Weekly_deaths.csv")
```

```python
deaths_df
```

```python
#review data types
deaths_df.dtypes
```

```python
#drop unneeded comlumns
deaths_df = deaths_df.drop(["Data As Of", "Start Date", "End Date", "Year", "Month", "MMWR Week", "Total Deaths", "Footnote", "Age Group"], axis=1)
```

```python
#drop unneeded rows
deaths_df = deaths_df.loc[(deaths_df["HHS Region"] != "United States") & (deaths_df["Group"] == "By Week") & (deaths_df ["Race and Hispanic Origin Group"] != "Unknown")]
```

```python
#drop group column
deaths_df = deaths_df.drop(["Group"], axis=1)
deaths_df
```

```python
deaths_df.dtypes
```

```python
#convert week to datetime
deaths_df["Week-Ending Date"] = pd.to_datetime(deaths_df["Week-Ending Date"])
```

```python
deaths_df
```

```python
#group data
weekly_data_df = deaths_df.groupby(["Week-Ending Date", "HHS Region", "Race and Hispanic Origin Group"]).sum()["COVID-19 Deaths"]
weekly_data_df.reset_index()
print(weekly_data_df)
```

```python
#export clean covid death data to CSV
weekly_data_df.to_csv("Resources\Clean_Data.csv")
```

```python
#read in vaccine data
JJ_df = pd.read_csv("Resources/Janssen.csv")
Moderna_df = pd.read_csv("Resources/Moderna.csv")
Pfizer_df = pd.read_csv("Resources/Pfizer.csv")
```

```python
JJ_df
```

```python
# create HHS region data
groups = {
    "1": ["Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island", "Vermont"],
    "2": ["New Jersey", "New York", "Puerto Rico", "Virgin Islands"],
    "3": ["Delaware", "District of Columbia", "Maryland", "Pennsylvania", "Virginia", "West Virginia"],
    "4": ["Alabama", "Florida", "Georgia", "Kentucky", "Mississippi", "North Carolina", "South Carolina", "Tennessee"],
    "5": ["Illinois", "Indiana", "Michigan", "Minnesota", "Ohio", "Wisconsin"],
    "6": ["Arkansas", "Louisiana", "New Mexico", "Oklahoma", "Texas"],
    "7": ["Iowa", "Kansas", "Missouri", "Nebraska"],
    "8": ["Colorado", "Montana", "North Dakota", "South Dakota", "Utah", "Wyoming"],
    "9": ["Arizona", "California", "Hawaii", "Nevada"],
    "10": ["Alaska", "Idaho", "Oregon", "Washington"]
}
```

```python
#clean JJ data
#add HHS region data to JJ DF
JJ_df["HHS Region"] = (
    JJ_df["Jurisdiction"]
    .apply(lambda x: [k for k in groups.keys() if x in groups[k]])
    .str[0]
    .fillna("Other")
)

#convert week to datetime
JJ_df["Week of Allocations"] = pd.to_datetime(JJ_df["Week of Allocations"])

#group data
clean_JJ = JJ_df.groupby(["HHS Region", "Week of Allocations"]).sum()["1st Dose Allocations"]
clean_JJ.reset_index()
print(clean_JJ)
```

```python
#clean Pfizer data
#add HHS region data to Pfizer DF
Pfizer_df["HHS Region"] = (
    Pfizer_df["Jurisdiction"]
    .apply(lambda x: [k for k in groups.keys() if x in groups[k]])
    .str[0]
    .fillna("Other")
)

#convert week to datetime
Pfizer_df["Week of Allocations"] = pd.to_datetime(Pfizer_df["Week of Allocations"])

#group data
clean_Pfizer = Pfizer_df.groupby(["HHS Region", "Week of Allocations"]).sum()["2nd Dose Allocations"]
clean_Pfizer.reset_index()
print(clean_Pfizer)
```

```python
#clean Moderna data
#add HHS region data to Moderna DF
Moderna_df["HHS Region"] = (
    Moderna_df["Jurisdiction"]
    .apply(lambda x: [k for k in groups.keys() if x in groups[k]])
    .str[0]
    .fillna("Other")
)

#convert week to datetime
Moderna_df["Week of Allocations"] = pd.to_datetime(Moderna_df["Week of Allocations"])

#group data
clean_Moderna = Moderna_df.groupby(["HHS Region", "Week of Allocations"]).sum()["2nd Dose Allocations"]
clean_Moderna.reset_index()
print(clean_Moderna)
```
